```python
import math
from collections import Counter

class MeraKNN:
    def __init__(self, neighbour):
        self.X_train = None
        self.y_train = None
        self.neighbour = neighbour

    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def predict(self, X_test):
        predictions = []

        for i in range(X_test.shape[0]):
            knnList = []

            # calculate distances
            for j in range(self.X_train.shape[0]):
                dist = math.dist(X_test[i], self.X_train[j])
                knnList.append((dist, j))

            # sort and select k nearest
            sorted_knn = sorted(knnList, key=lambda x: x[0])[:self.neighbour]
            indices = [index for (_, index) in sorted_knn]

            # get corresponding labels
            outputs = [self.y_train[idx] for idx in indices]

            # majority vote
            most_common = Counter(outputs).most_common(1)[0][0]
            predictions.append(most_common)

        return predictions
```
